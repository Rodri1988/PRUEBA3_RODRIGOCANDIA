```python
from IPython.core.interactiveshell import re
def verificar_dni(dni):
    dni = dni.replace("-", "")
    dni_numero = dni[:-1]
    verificador = dni[-1].upper()
#rut:16787945-4= tiene 8 mas un numero verificador
#DNI:30.686.957-X tiene 8 mas un numero verificador
    suma = 0
    multiplicador = 2
    for digito in reversed(dni_numero):
        suma += int(digito) * multiplicador
        multiplicador += 1
        if multiplicador > 7:
            multiplicador = 2

    resto = suma % 11
    digito_verificador = str(11 - resto)

    if digito_verificador == "10":
        digito_verificador = "K"
    elif digito_verificador == "11":
        digito_verificador = "0"

    return digito_verificador == verificador


def grabar_persona(personas):
    dni = input("Ingrese el DNI de la persona sin puntos (con guión): ")
    if not verificar_dni(dni):
        print("DNI incorrecto")
        return

    nombre = input("Ingrese el nombre completo de la persona: ")
    if len(nombre) < 8:
        print("Nombre demasiado corto")
        return

    estado = input("Ingrese su Estado Civil: ")
    if  len(estado) < 5:
        print("no corresponde estado")
        return

    edad = int(input("Ingrese la edad de la persona: "))
    if edad < 0:
        print("Edad inválida")
        return

    pais_nacimiento = input("Ingrese el país de nacimiento: ")
    ciudad_nacimiento = input("Ingrese la ciudad de nacimiento: ")

    persona = [dni, nombre,estado, edad, pais_nacimiento, ciudad_nacimiento]
    personas.append(persona)

    print("Datos de la persona guardados exitosamente")


def buscar_persona(personas):
    dni = input("Ingrese el DNI de la persona a buscar (con guión): ")

    if dni.endswith("-X"):
        dni = dni[:-2]

    for persona in personas:
        if persona[0] == dni:
            print("El DNI es válido")
            print("Nombre:", persona[1])
            print("Edad:", persona[3])
            print("Estado:",persona[2])
            print("País de nacimiento:", persona[4])
            print("Ciudad de nacimiento:", persona[5])

            if persona[4].lower() == "Argentina":
                print("La persona pertenece a Argentina")

            return

    print("DNI no encontrado")


def imprimir_certificados(personas):


    dni = input("Ingrese el DNI  de la persona: ")

    for persona in personas:
        if persona[0] == dni:
            print("Certificado:")
            print("Nombre:", persona[1])
            print("Edad:", persona[3])
            print("DNI:", persona[0])
            print("Pais de nacimiento:",persona[4])
            print("Ciudad de Nacimiento:", persona[5])
            print("Estado Civil:", persona[2])


            return

    print("DNI no encontrado")


def eliminar_persona(personas):
    dni = input("Ingrese el DNI de la persona a eliminar (con guión): ")

    for persona in personas:
        if persona[0] == dni:
            personas.remove(persona)
            print("Persona eliminada exitosamente")
            return

    print("DNI no encontrado")


def main():
    personas = []

    print("Bienvenido al Servicio de RENAPER")

    while True:
        print("\n--- Opciones ---")
        print("1. Grabar")
        print("2. Buscar")
        print("3. Imprimir certificados")
        print("4. Eliminar")
        print("5. Salir")

        opcion = input("Ingrese la opción deseada: ")

        if opcion == "1":
            grabar_persona(personas)
        elif opcion == "2":
            buscar_persona(personas)
        elif opcion == "3":
            imprimir_certificados(personas)
        elif opcion == "4":
            eliminar_persona(personas)
        elif opcion == "5":
            print("¡Hasta luego!")
            break
        else:
            print("Opción inválida. Intente nuevamente.")

if __name__ == "__main__":
    main()
```
